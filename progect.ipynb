```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.columns
```

# 2 Data Cleaning

```python
data = data.drop("Unnamed: 0" , axis = 1)
data.head()
```

## 2.1 Changing data types

```python
data["player_agent"] = data["player_agent"].str.replace('\.\.\.$', '', regex=True);
```

```python
data.dtypes
```

## 2.2 Making Additional DataFrames

```python
data.isna().sum()
```

```python
data_contract = data.dropna(subset=["contract_expires"], axis=0)
data_contract["contract_expires"].isna().sum()
data_contract["contract_expires"].unique()
data_contract["contract_expires"] = data_contract["contract_expires"].replace('nan', np.nan)
data_contract = data_contract.dropna(subset=["contract_expires"], axis=0)
```

```python
data_price = data.dropna(subset=["price"], axis=0)
data_price["price"].isna().sum()
```

```python
data_youngers = data_price[data_price['club'].str.endswith(('U19', 'II', 'U17'))]
data_youngers
```

```python
data_prof = data_price[~data_price['club'].str.endswith(('U19', 'II', 'U17'))]
data_prof
```

```python
data.info()
```
